# ---------------- DICTIONARY QUESTIONS ----------------

```python
# 1. Create a dictionary with student name and marks
student = {"Rahul": 85, "Aman": 90, "Priya": 88}
print("1.", student)
```

```python
# 2. Access a value using key
print("2.", student["Rahul"])
```

```python
# 3. Add a new key-value pair into dictionary
student["Sneha"] = 95
print("3.", student)
```

```python
# 4. Update an existing value in dictionary
student["Aman"] = 92
print("4.", student)
```

```python
# 5. Remove a key using pop()
student.pop("Priya")
print("5.", student)
```

```python
# 6. Remove the last inserted item using popitem()
student.popitem()
print("6.", student)
```

```python
# 7. Display all keys from dictionary
print("7.", student.keys())
```

```python
# 8. Display all values from dictionary
print("8.", student.values())
```

```python
# 9. Display all items from dictionary
print("9.", student.items())

```

```python
# 10. Check whether a key exists in dictionary
if "Rahul" in student:
    print("10. Key exists")
else:
    print("10. Key does not exist")
```

```python
# 11. Find length of dictionary
print("11. Length =", len(student))
```

```python
# 12. Copy one dictionary into another
copy_dict = student.copy()
print("12.", copy_dict)
```

```python
# 13. Clear all items from dictionary
copy_dict.clear()
print("13.", copy_dict)
```

```python
# 14. Create dictionary using fromkeys()
keys = ("a", "b", "c")
new_dict = dict.fromkeys(keys, 0)
print("14.", new_dict)
```

```python
# 15. Get value of a key using get() method
print("15.", student.get("Aman"))
```

```python
# 16. Iterate through dictionary using loop
print("16.")
for key, value in student.items():
    print(key, ":", value)
```

```python
# 17. Merge two dictionaries
d1 = {"x": 1, "y": 2}
d2 = {"z": 3}
d1.update(d2)
print("17.", d1)
```

```python
# 18. Create nested dictionary and access inner values
nested = {
    "student1": {"name": "Rahul", "marks": 85},
    "student2": {"name": "Aman", "marks": 90}
}
print("18.", nested["student1"]["name"])
```

```python
# 19. Convert dictionary keys into list
keys_list = list(student.keys())
print("19.", keys_list)
```

```python
# 20. Sort dictionary keys and display them
sorted_keys = sorted(student.keys())
print("20.", sorted_keys)
```
